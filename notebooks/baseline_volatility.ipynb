# Civic Lens: Baseline Volatility Review

Phase 9 plausibility notebook for the current processed outputs.

- `training_metrics.csv` contains ward and borough rows for 2014->2018
- `backtest_actuals_2022.csv` is borough-only under the current Phase 7 concordance
- `party_swings_training.csv` and `party_swings_backtest.csv` provide the party-level swing details used downstream


```python
import pandas as pd

tm = pd.read_csv('../data/processed/training_metrics.csv')
ba = pd.read_csv('../data/processed/backtest_actuals_2022.csv')
pst = pd.read_csv('../data/processed/party_swings_training.csv')
psb = pd.read_csv('../data/processed/party_swings_backtest.csv')

print('training rows:', len(tm))
print(tm['computation_level'].value_counts().to_string())
print('\nbacktest rows:', len(ba))
print(ba['computation_level'].value_counts().to_string())
```

```python
for name, df in [('training', tm), ('backtest', ba)]:
    vol = df['volatility_score'].dropna()
    print(f'\n{name} volatility summary')
    print(vol.describe().round(3).to_string())
    fi_cols = [c for c in df.columns if c.startswith('fi_')]
    for col in fi_cols:
        print(f'{col}: min={df[col].dropna().min():.3f} max={df[col].dropna().max():.3f}')
```

```python
print('Top training ward volatility rows')
display(tm.sort_values('volatility_score', ascending=False).head(10))

print('Top borough backtest volatility rows')
display(ba.sort_values('volatility_score', ascending=False).head(10))
```
